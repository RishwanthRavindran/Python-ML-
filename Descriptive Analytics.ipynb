## Grouping and Aggregating

```python
DiamondData.groupby('cut')['price'].mean()
```

```python
price_by_cut = DiamondData.groupby('cut')['price'].mean().reset_index()
print(price_by_cut)
```

```python
price_by_cut_color = DiamondData.groupby(['cut','color'])['price'].mean().reset_index()
print(price_by_cut_color)
```

## Joining DataFrames

```python
price_comparison = price_by_cut_color.merge(price_by_cut,on ='cut',how ='outer')
```

```python
price_comparison
```

## Re-naming Columns

```python
price_comparison.rename(columns = {'price_x': 'price_by_cut_color', 'price_y':'price_by_cut'},inplace = True)
price_comparison
```

## Applying Operations to Multiple Columns

```python
price_comparison['change'] = price_comparison.apply(lambda ch: 
                                                    (ch.price_by_cut_color - ch.price_by_cut)/ch.price_by_cut, axis = 1)
                        
```

```python
price_comparison
```
